Мы создадим модель логистической регрессии для классификации
ирисов. Мы воспользуемся набором данных ирисы Фишера - это очень популярный
датасет, который описывает четыре типа ирисов по четырем характеристикам.


```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt



iris = load_iris()

x = iris.data[:, :2]
y = iris.target


clf = LogisticRegression()

clf.fit(x, y)

y_pred = clf.predict(x)

plt.figure(figsize=(10, 6))
plt.scatter(x[:, 0], x[:, 1], c=y_pred, cmap=plt.cm.Set1, edgecolor='k')

plt.xlabel('Lenght')
plt.ylabel('width')
plt.show()
```
