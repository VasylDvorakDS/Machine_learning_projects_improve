```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 2, 3, 4, 5])
y1 = np.array([1, 2, 3, 4, 5])
y2 = np.array([1, 1, 4, 9, 0])
```

```python
plt.scatter(x, y, s=30)
plt.plot(x, y1, 'g')
plt.plot(x, y2, 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
error1 = sum(y1 - y)
error2 = sum(y2 - y)
print(error1)
print(error2)
```

```python
error1 = sum((y1 - y) ** 2)
error2 = sum((y2 - y) ** 2)
print(error1)
print(error2)
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(0)
x = np.random.rand(1000, 1)
y = 1 + 4*x + np.random.randn(1000, 1)

d = {'x': x.reshape(1, 1000)[0], 'y': y.reshape(1, 1000)[0]}
df = pd.DataFrame(data=d)
df.head()
```

```python
plt.scatter(df['x'], df['y'], s=10)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
train, test = train_test_split(df, train_size=0.7, random_state=123)
```

```python
model = LinearRegression()
model.fit(train[['x']], train[['y']])
```

```python
print('w0 ', model.intercept_)
print('w1 ', model.coef_)
```

```python
y = 1.08 + 3.84 * x
```

```python
y_predict_train = model.predict(train[['x']])
print(y_predict_train[:, 0][:10])
```

```python
y_predict_train = model.intercept_ + model.coef_ * train[['x']]
print(y_predict_train[:10])
```

```python
plt.plot(train['x'], y_predict_train, linewidth=4, c='g')
plt.scatter(train['x'], train['y'], s=10, c='b')
plt.scatter(test['x'], test['y'], s=10, c='r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import pandas
from pandas import DataFrame
```

```python
data = pandas.read_csv('123.csv')
```

```python
X = DataFrame(data,columns=['production_budget_usd'])
y = DataFrame(data,columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)

plt.plot(X, model.predict(X), color='red', linewidth=3)
plt.title('Film Cost VS Global Revenue')
plt.xlabel('Production Budget')
plt.ylabel('Worldwide Gross')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```
