```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [0, 1, 2, 3, 4, 5, 6]
y = [0, 1, 2, 3, 4, 5, 6]
```

```python
plt.scatter(y, y, s=10)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
y1 = np.array([0, 1, 2, 3, 4, 5, 6])
y2 = np.array([0, 1, 2, 10, 5, 4, 8])
y3 = np.array([0, 1, 3, 1, 3, 4, 8])

plt.scatter(y, y, s=10)
plt.plot(y1, 'g')
plt.plot(y2, 'r')
plt.plot(y3, 'b')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
t1 = sum(y - y1)
t2 = sum(y - y2)
t3 = sum(y - y3)

print(t1)
print(t2)
print(t3)
```

```python
t1 = sum(y - y1) / len(y)
t2 = sum(y - y2) / len(y)
t3 = sum(y - y3) / len(y)

print(t1)
print(t2)
print(t3)
```

```python
t1 = sum(abs(y - y1)) / len(y)
t2 = sum(abs(y - y2)) / len(y)
t3 = sum(abs(y - y3)) / len(y)

print(t1)
print(t2)
print(t3)
```
